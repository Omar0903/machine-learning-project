# Check Outliers of data

```python
dataframe1['Number of Employees'] = pd.to_numeric(dataframe1['Number of Employees'], errors='coerce').astype('float64')
dataframe1['IPO'] = pd.to_numeric(dataframe1['IPO'], errors='coerce').astype('float64')
columns = ['Year Founded', 'Number of Employees','Number of Acquisitions','Total Funding ($)','Number of Employees (year of last update)']

for i in columns:
    q1 = np.percentile(dataframe1[i], 25)
    q3 = np.percentile(dataframe1[i], 75)
    norm_range = (q3 - q1) * 1.5
    # Identify lower outliers
    lower_outliers = dataframe1[dataframe1[i] < (q1 - norm_range)]
    # Identify upper outliers
    upper_outliers = dataframe1[dataframe1[i] > (q3 + norm_range)]
    # Count the total number of outliers
    outliers = len(lower_outliers) + len(upper_outliers)
    print(f"The number of outliers in {i}: {outliers}")
    # Replace outliers with adjusted values
    dataframe1[i] = np.where(dataframe1[i] < (q1 - norm_range), q1 - norm_range, dataframe1[i])
    dataframe1[i] = np.where(dataframe1[i] > (q3 + norm_range), q3 + norm_range, dataframe1[i])
```

# Remove duplicated rows

```python
dataframe1 = dataframe1.duplicated()
```

#Dealing with nulls in data

```python
#Replace NULL with mean values
dataframe1['Number of Employees'].fillna(dataframe1['Number of Employees'].mean(),inplace=True)
dataframe1['Number of Employees (year of last update)'].fillna(dataframe1['Number of Employees (year of last update)'].median(),inplace=True)
dataframe1["IPO"] = dataframe1["IPO"].replace("Not yet", 0)
dataframe1["IPO"] = pd.to_numeric(dataframe1["IPO"], errors='coerce')
dataframe1["IPO"].fillna(dataframe1["IPO"].mean(), inplace=True)
dataframe1["IPO"] = np.floor(dataframe1["IPO"])
#Replace NULL with mode values
dataframe1["State / Region (HQ)"].fillna(dataframe1["State / Region (HQ)"].mode()[0],inplace=True)
dataframe1["City (HQ)"].fillna(dataframe1["City (HQ)"].mode()[0],inplace=True)
#Replace NULL with unknown
dataframe1["Twitter"] = dataframe1["Twitter"].fillna("Unknown")
dataframe1["Board Members"] = dataframe1["Board Members"].fillna("Unknown")
```
